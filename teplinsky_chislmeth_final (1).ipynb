используемые библиотеки:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # plt - построение графиков
import seaborn as sns            # sns - стиль графиков

sns.set_theme(style="whitegrid", palette="deep")
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
# table - вывод таблиц
def table(df):
    return (
        df.style
        .hide(axis="index")
        .set_table_styles([
            {"selector": "th", "props": [("background-color", "#5f6368"),
                                         ("color", "white"),
                                         ("font-size", "11pt"),
                                         ("text-align", "center")]},
            {"selector": "td", "props": [("font-size", "10.5pt"),
                                         ("text-align", "center")]},
        ])
    )

# roots_on_grid - точки пересечения по сетке
def roots_on_grid(x, y):
    roots = []
    for i in range(len(x) - 1):
        if np.isnan(y[i]) or np.isnan(y[i + 1]):
            continue
        if y[i] == 0:
            roots.append(float(x[i]))
        elif y[i] * y[i + 1] < 0:
            roots.append(float((x[i] + x[i + 1]) / 2))
    return roots

# scan - отрезки со сменой знака
def scan(f, a, b, h):
    x_step = np.arange(a, b + h, h)
    y_step = np.array([f(xi) for xi in x_step], dtype=float)
    intervals = []  # найденные отрезки
    for i in range(len(x_step) - 1):  # соседние точки сетки
        if np.isnan(y_step[i]) or np.isnan(y_step[i + 1]):
            continue
        if y_step[i] == 0 or y_step[i] * y_step[i + 1] < 0:  # ноль или смена знака
            intervals.append((x_step[i], x_step[i + 1]))
    return intervals

# show_intervals - вывод ответа
def show_intervals(intervals, digits=1):
    parts = []
    for a, b in intervals:
        a = round(float(a), digits)
        b = round(float(b), digits)
        parts.append(f"[{a}; {b}]")
    return ", ".join(parts).replace("-0.0", "0").replace("0.0", "0")
```

## 2.1
### упражнение 1

а) \(\lg x + 6 = x^2\)


```python
f = lambda x: np.log10(x) + 6 - x**2

x = np.linspace(1e-6, 4, 2500)
y = f(x)

x_pair = np.linspace(1e-6, 4, 2500)
y1 = np.log10(x_pair) + 6
y2 = x_pair**2

intervals = scan(f, 1e-6, 4, 0.1)
roots = roots_on_grid(x_pair, y1 - y2)

# axes - два графика в одной фигуре
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# plot - линия функции
axes[0].plot(x, y, linewidth=2, label="lg(x)+6-x^2")
axes[0].axhline(0, color="black", linewidth=1)  # ox
axes[0].axvline(0, color="black", linewidth=1)  # oy
axes[0].set_xlim(0, 4)
axes[0].set_ylim(-1, 8)
axes[0].set_title("f(x) = 0")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].legend()

axes[1].plot(x_pair, y1, linewidth=2, label="lg(x)+6")
axes[1].plot(x_pair, y2, linewidth=2, label="x^2")

for xi in roots:
    yi = np.interp(xi, x_pair, y1)
    axes[1].scatter(xi, yi, color="black", s=20, zorder=5)
    axes[1].vlines(xi, 0, yi, colors="red", linestyles="dashed", linewidth=1.2)

axes[1].axhline(0, color="black", linewidth=1)
axes[1].axvline(0, color="black", linewidth=1)
axes[1].set_xlim(0, 4)
axes[1].set_ylim(-1, 8)
axes[1].set_title("пересечение y1 и y2")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].legend()

plt.tight_layout()
plt.show()

print("ответ:", show_intervals(intervals))
```

б) \(x\sin x - 1 = 0\)


```python
f = lambda x: x * np.sin(x) - 1

x_fx = np.linspace(-10, 10, 3000)
y_fx = f(x_fx)

x_sin = np.linspace(-10, 10, 3000)  # x для sin(x)
x_left = np.linspace(-10, -0.1, 1500)
x_right = np.linspace(0.1, 10, 1500)

y_sin = np.sin(x_sin)  # sin(x)
y_left = 1 / x_left
y_right = 1 / x_right

intervals = scan(f, -10, 10, 0.1)
roots = roots_on_grid(x_left, np.sin(x_left) - 1 / x_left) + roots_on_grid(x_right, np.sin(x_right) - 1 / x_right)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(x_fx, y_fx, linewidth=2, label="x sin(x)-1")
axes[0].axhline(0, color="black", linewidth=1)
axes[0].axvline(0, color="black", linewidth=1)
axes[0].set_xlim(-10, 10)
axes[0].set_ylim(-10, 10)
axes[0].set_title("f(x) = 0")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].legend()

axes[1].plot(x_sin, y_sin, linewidth=2, label="sin(x)")
axes[1].plot(x_left, y_left, linewidth=2, label="1/x")
axes[1].plot(x_right, y_right, linewidth=2)

for xi in roots:
    yi = np.sin(xi)
    axes[1].scatter(xi, yi, color="black", s=20, zorder=5)
    axes[1].vlines(xi, 0, yi, colors="red", linestyles="dashed", linewidth=1.1)

axes[1].axhline(0, color="black", linewidth=1)
axes[1].axvline(0, color="black", linewidth=1)
axes[1].set_xlim(-10, 10)
axes[1].set_ylim(-2, 2)
axes[1].set_title("пересечение y1 и y2")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].legend()

plt.tight_layout()
plt.show()

print("ответ:", show_intervals(intervals))
```

## 2.2
### упражнение


```python
# f - исходная функция
def f(x):
    return x * np.sin(x) - 1

a = 1.1
b = 1.2
eps = 1e-4
rows = []  # строки таблицы
n = 0

while (b - a) / 2 > eps:
    fa = f(a)
    fb = f(b)
    c = (a + b) / 2  # середина
    fc = f(c)

    rows.append([
        n,
        round(a, 6),
        round(b, 6),
        round(fa, 6),
        round(fb, 6),
        round(c, 6),
        round((b - a) / 2, 6)
    ])

    if fa * fc <= 0:  # корень слева
        b = c
    else:
        a = c

    n += 1

c = (a + b) / 2
rows.append([
    n,
    round(a, 6),
    round(b, 6),
    round(f(a), 6),
    round(f(b), 6),
    round(c, 6),
    round((b - a) / 2, 6)
])

df = pd.DataFrame(rows, columns=["n", "a", "b", "f(a)", "f(b)", "c", "Δ"])
display(table(df))

print(f"ответ: x ≈ {c:.6f}")
```

### упражнение 2

\(x\sin x - 1 = 0\), метод простой итерации, \(\varepsilon = 10^{-5}\)


```python
# F - уравнение
def F(x):
    return x * np.sin(x) - 1

# phi - шаг итерации
def phi(x):
    return x - 0.5 * F(x)

eps = 1e-5
x_prev = 1.0
rows = []

for n in range(1, 51):
    x_next = phi(x_prev)
    d = abs(x_next - x_prev)

    rows.append([
        n,
        round(x_prev, 6),
        round(x_next, 6),
        round(d, 6),
        round(F(x_prev), 6)
    ])

    if d < eps:
        break

    x_prev = x_next

df = pd.DataFrame(rows, columns=["n", "x_n", "x_(n+1)", "|x_(n+1)-x_n|", "F(x_n)"])
display(table(df))

print(f"ответ: x ≈ {x_next:.6f}")
```

### упражнение 2

\(x^3 + 3x^2 - 24x + 1 = 0\), комбинированный метод, \([0; 1]\), \(
arepsilon = 10^{-6}\)


```python
# F - уравнение
def F(x):
    return x**3 + 3 * x**2 - 24 * x + 1

# dF - производная
def dF(x):
    return 3 * x**2 + 6 * x - 24

# rows - строки таблицы
def combined_method_table(a, b, eps=1e-6):
    if F(a) * F(b) >= 0:
        print("на отрезке нет корня или их четное число")
        return None

    if F(a) * (6 * a + 6) > 0:  # F''(x) = 6x + 6
        x_k, x_h = a, b
    else:
        x_k, x_h = b, a

    rows = []
    for n in range(1, 100):
        x_k_new = x_k - F(x_k) / dF(x_k)  # касательные
        x_h_new = x_h - F(x_h) * (x_h - x_k) / (F(x_h) - F(x_k))  # хорды
        err = abs(x_k_new - x_h_new)

        rows.append([
            n,
            round(x_k_new, 6),
            round(x_h_new, 6),
            round(err, 6)
        ])

        if err < eps:
            break

        x_k, x_h = x_k_new, x_h_new

    df = pd.DataFrame(rows, columns=["n", "x (касат.)", "x (хорды)", "|x_k - x_h|"])
    display(table(df))
    print(f"ответ: x ≈ {(x_k_new + x_h_new) / 2:.6f}")

combined_method_table(0, 1)
```

```python
x = np.linspace(-1, 4, 1500)
y1 = x**2 - 3.2 * x
y2 = np.ones_like(x)
roots = roots_on_grid(x, y1 - y2)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, linewidth=2, label="x^2 - 3.2x")
plt.plot(x, y2, linewidth=2, label="1")

for xi in roots:
    yi = np.interp(xi, x, y1)
    plt.scatter(xi, yi, color="black", s=20, zorder=5)
    plt.vlines(xi, 0, yi, colors="red", linestyles="dashed", linewidth=1.1)

plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1)

plt.xlim(-1, 4)
plt.ylim(-2, 3)
plt.title("графическое отделение корней", fontsize=14)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()

print("ответ:", "[-1; 0], [3; 4]")

# phi - правая часть
def phi(x):
    return np.sqrt(3.2 * x + 1)

x_prev = 3.5
eps = 1e-5
rows = []

for n in range(1, 51):
    x_next = phi(x_prev)
    d = abs(x_next - x_prev)

    rows.append([
        n,
        round(x_prev, 6),
        round(x_next, 6),
        round(d, 6)
    ])

    if d < eps:
        break

    x_prev = x_next

df = pd.DataFrame(rows, columns=["n", "x_n", "x_(n+1)", "|x_(n+1)-x_n|"])
display(table(df))

print(f"один корень: x ≈ {x_next:.4f}")
```

### 4 задание

комбинированный метод хорд и касательных, \(\varepsilon = 10^{-6}\)

```python
def F(x):
    return 3 * np.sin(8 * x) - 0.7 * x + 0.9

def dF(x):
    return 24 * np.cos(8 * x) - 0.7

def ddF(x):
    return -192 * np.sin(8 * x)

a = -0.1
b = -0.03
eps = 1e-6

# старт для касательных и хорд
if F(a) * ddF(a) > 0:
    xt = a
    xh = b
    c = a
else:
    xt = b
    xh = a
    c = b

rows = []  # строки таблицы

for n in range(1, 50):
    xt1 = xt - F(xt) / dF(xt)  # шаг касательных
    xh1 = (c * F(xh) - xh * F(c)) / (F(xh) - F(c))  # шаг хорд

    delta = abs(xt1 - xh1) / 2  # оценка погрешности
    mid = (xt1 + xh1) / 2       # приближение корня

    rows.append([
        n,
        round(float(xt), 7),
        round(float(xh), 7),
        round(float(xt1), 7),
        round(float(xh1), 7),
        round(float(mid), 7),
        round(float(delta), 7)
    ])

    if delta < eps:
        break

    xt = xt1
    xh = xh1

df = pd.DataFrame(rows, columns=["n", "x_кас", "x_хорд", "x_кас_нов", "x_хорд_нов", "x", "Δ"])
display(table(df))

print(f"ответ: x ≈ {mid:.7f}")
```
